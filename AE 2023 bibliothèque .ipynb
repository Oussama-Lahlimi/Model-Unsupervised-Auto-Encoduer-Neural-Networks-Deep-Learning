```python
from sklearn.datasets import make_regression
#Genarate a regression data
#generate the data
X,Y = make_regression(n_samples=1000,n_features=150,n_informative=10,noise=0.1,random_state=1)
```

```python
#processin
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
#splitting to traine end test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.34,random_state=1)

process=MinMaxScaler()
process.fit(X_train) 
X_train=process.transform(X_train)
X_test=process.transform(X_test)
```

```python
#builde the model AE : Encoder and Decoder
#Encoder 
#affichege seulment les donn√©es entre 1 et 0 max,min 
# X_train.shape
# sans shape 
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense , ReLU, BatchNormalization
inputs= X_test.shape[1]
#define the Encoder
encoder_input =Input(shape=(inputs,))
encoder_layer=Dense(inputs*2)(encoder_input)
encoder_layer=BatchNormalization()(encoder_layer)
encoder_layer=ReLU()(encoder_layer)

# define the end of the encoder ::
n_bottleNeck= inputs
bottleNeck =Dense(n_bottleNeck)(encoder_layer)


```

```python
#Decoder
decoder=Dense(inputs*2)(encoder_input)
decoder=BatchNormalization()(decoder)
decoder=ReLU()(decoder)
decoder_output=Dense(inputs,activation="linear")(decoder)

#define the AE

autoencoder =Model(inputs=encoder_input,outputs=decoder_output)



```

```python
#comilling end configuration
#autoencoder.compile(optimizers='adam',loss='mse')
autoencoder.compile(optimizer='adam', loss='mse')
```

```python
#Plot the AE
from tensorflow.keras.utils import plot_model
plot_model(autoencoder,'auto.jpg',show_shapes=True)
```

```python
#Train the autoencoder
#AE kayb9a yt3alam mn exp li dazo 3lih
history = autoencoder.fit(X_train,X_train,epochs=500,validation_data=(X_test,X_test))
#regression
#les erreur recoit petin dans aoutp par contre input
```

```python
from matplotlib import pyplot

pyplot.title("the losses")
pyplot.plot(history['loss'],label='train')
pyplot.plot(history.history['val_loss'],label='test')
pyplot.legend()
pyplot.show()

```
